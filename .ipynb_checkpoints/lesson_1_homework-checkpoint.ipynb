# Training the model for Kaggle Data

Resseting the path for the kaggle data

```python
path="data/dogscats-redux/"
```

## Creating and running the model

```python
batch_size = 64
vgg = Vgg16()
```

Getting the batches

```python
batches = vgg.get_batches(path=path+"train/", batch_size=batch_size)
valid_batches = vgg.get_batches(path=path+"valid/", batch_size=batch_size)
```

Finetuning and fitting our model

```python
vgg.finetune(batches)
vgg.fit(batches, valid_batches, nb_epoch=1)
```

```python
path="data/dogscats-redux/sample/"
test_batches = vgg.get_batches(path+"test", batch_size=4, class_mode=None)
```

```python
imgs = next(test_batches)
```

```python
filenames=test_batches.filenames[:4]
prediction = vgg.predict(imgs, False)
prob = prediction[0]
labels = prediction[1]
for i in len(prob):
    if labels[i] == 0:
        print(filenames[i] + ",{0}",1.0 - prob[i])
    else:
        print(filenames[i] + ",{0}prob[i])
```
